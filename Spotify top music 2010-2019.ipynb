```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('top10s.csv',encoding = "ISO-8859-1",index_col=0)
df_=df.copy()
df.columns=['title','artist','genre','year','beat_per_min','energy','danceability','loudness','liveness','valence','length','acousticness','speechiness','popularity']
df.columns[4]
df.columns[13]
```

```python
# draw the trend of characteristics
year=2010
result=[]
while year<2020:
    for col in df.columns[4:]:
        avg = df[df.year==year][col].mean()
        result.append(avg)
    year+=1

dict_chars={}
counter=4
start=0

while counter<14:
    char=df.columns[counter]
    dict_chars[char]=result[start::10]
    start+=1
    counter+=1
df_chars=pd.DataFrame(dict_chars,index=[year for year in range(2010,2020)])
# for col in df.columns[4:]:
#     df_chars.plot(y=col)
df_chars.plot(subplots=True,figsize=(10,100))
# # df_chars=pd.DataFrame(result)
# df_chars
```

```python
def top_10_most_popular_artists(df):
    print('top ten most popular artists in 2010s and the occurences of their songs among spotify\'s most popular:  ')
    df_artist = df['artist'].value_counts()[:10].to_frame()
    df_artist.columns = ['popular hits occurrences']
    print(df_artist)
    return df_artist

for artist in top_10_most_popular_artists(df).index.values:
    print('\n' + str(artist) + '\'s top 5 songs are: \n')
    counter=1
    for song in df[df.artist==artist].sort_values(by='popularity',ascending=False)[:5].title.values:
        print(str(counter) +'   ' + song)
        counter+=1
```
